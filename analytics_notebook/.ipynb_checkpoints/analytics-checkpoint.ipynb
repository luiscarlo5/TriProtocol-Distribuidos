## Leitura de Bibliotecas

```python
import pandas as pd
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leitura de arquivos CSV

```python
df_string = pd.read_csv("../data_analytics/resultados_tempo_string.csv")
df_json = pd.read_csv("../data_analytics/resultados_tempo_json.csv")
df_protobuf = pd.read_csv("../data_analytics/resultados_tempo_protobuf.csv")

df_todos = [df_string, df_json, df_protobuf] 
```

```python
# exemplo com operação de soma nos dados do protocolo buffer
df_protobuf[df_protobuf['operacao'] == 'soma' ]
```

```python
df_protobuf.shape
```

### Dados de Média e Desvio Padrão

```python
nomes = ["String", "JSON", "Protobuf"]

for tipo in ['echo', 'soma', 'status', 'historico', 'timestamp']:
    print(f"\nOperação {tipo.upper()}")
    for nome, df_main in zip(nomes, df_todos):
        df_aux = df_main[df_todos[0]['operacao'] == tipo ]

        print(f"Protocolo {nome}: Média {round(df_aux['tempo_resposta'].mean(), 3)}")
```

```python
nomes = ["String", "JSON", "Protobuf"]
operacoes = ['echo', 'soma', 'status', 'historico', 'timestamp']

resultados = []

for tipo in operacoes:
    for nome, df_main in zip(nomes, df_todos):

        # Filtrar operação no DF correto
        df_aux = df_main[df_main['operacao'] == tipo]

        resultados.append({
            "operacao": tipo,
            "protocolo": nome,
            "media": df_aux["tempo_resposta"].mean(),
            "desvio": df_aux["tempo_resposta"].std()
        })

df_resultado = pd.DataFrame(resultados)
print(df_resultado)
```

## Análise Gráfica
> Paleta de Cores para 3 cores Azuis

```python
paleta_azul = sns.color_palette("Blues", n_colors=3)
```

### Comparação da Média entre diferentes Protocolos

```python
sns.set(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(10,5))
sns.barplot(
    data=df_resultado,
    x="operacao",
    y="media",
    hue="protocolo",
    palette=paleta_azul,
    edgecolor="black"
)
plt.title("Média do tempo_resposta por operação e protocolo")
plt.ylabel("Média (s)")
plt.xlabel("Operação")
plt.legend(title="Protocolo")
plt.tight_layout()
plt.show()
```

### Comparação de Desvio Padrão entre diferentes Protocolos

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=df_resultado,
    x="operacao",
    y="desvio",
    hue="protocolo",
    palette=paleta_azul,
    edgecolor="black"
)
plt.title("Desvio padrão do tempo_resposta por operação e protocolo")
plt.ylabel("Desvio padrão")
plt.xlabel("Operação")
plt.legend(title="Protocolo")
plt.tight_layout()
plt.show()
```
